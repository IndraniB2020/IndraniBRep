```python
def meal_type(df1):
    
    if(df1['food_item'] == 'Black Sugar Bubble Tea'):
        return 'Dessert'
    elif (df1['food_item'] == 'Braised Beef Noodles'):
        return 'Dinner'
    elif (df1['food_item'] == 'Cha Siu Bao'):
        return 'Snacks'
    elif (df1['food_item'] == 'Gam Ja Tang'):
        return 'Dinner'
    elif (df1['food_item'] == 'Herbal Grass Jelly Drink'):
        return 'Beverages'
    elif (df1['food_item'] == 'Matcha Cheese Cake'):
        return 'Dessert'
    elif (df1['food_item'] == 'Passionfruit Green Tea'):
        return 'Beverages'
    elif (df1['food_item'] == 'Pork Belly Bao'):
        return 'Snacks'
    elif (df1['food_item'] == 'Red Bean & Taro Slush'):
        return 'Beverages'

        
df1['meal_type'] = df1.apply(meal_type, axis = 1)
```

```python
from dateutil.parser import parse
#df['Time_diff'] =   df['delivered_at'] - df['ordered_at']
  
df['delivered_at'] = pd.to_datetime(df['delivered_at'], format='%m/%d/%Y %H:%M')
df['ordered_at'] = pd.to_datetime(df['ordered_at'], format='%m/%d/%Y %H:%M')
#df['Time_diff'] =   df['delivered_at'] - df['ordered_at']

## difference of time
DIFF_Time_spend = df['delivered_at']  - df['ordered_at']
##DIFF_Time_spend  ## total time difference
df['Time_spend_mins'] = DIFF_Time_spend.dt.components.minutes
df
```

```python
deliver_sort =  df.groupby(['delivered_to'])['id'].count().sort_values()
df.delivered_to.nunique() ## 992 unqiue delivery sites
df.sort_values(by=['delivered_to'])
df_refined = df[['id','delivered_to','Time_spend_mins']]
df_refined
```

```python
df_refined.groupby('delivered_to')['id'].count().sort_values()
```

```python
df_refined['Delivery_freq'] = df_refined.groupby('delivered_to')['id'].transform('count')
#df_refined
df_refined.sort_values(by=['Delivery_freq'])
df_refined
```

```python
df2 = df2.rename(columns={'order_id':'id'})
df2
```

```python
df3 = pd.merge(df_refined, df2, on=['id'], how='inner')
df4 = df3.sort_values(by=['Delivery_freq'])
df4.columns
```

```python
df4.dtypes
```

```python
df4 = df4.fillna('0')
df4['Beverages'] = df4['Beverages'].apply(np.int64)
df4['Dessert'] = df4['Dessert'].apply(np.int64)
df4['Dinner'] = df4['Dinner'].apply(np.int64)
df4['Snacks'] = df4['Snacks'].apply(np.int64)
```

```python
df4.dtypes
df4['Customer_status'] = df4['Delivery_freq'].apply(lambda x: 'New' if x == 1 else 'old')
```

```python
df4.to_csv('refined.csv', index=False)
```
