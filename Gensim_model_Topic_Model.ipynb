```python
df['Description'] = df['Description'].str.replace('\d+','')
df['Description']
```

```python
#remove punctuation
def remove_pun(txt):
    txt_nopun = "".join([c for c in txt if c not in string.punctuation])
    return txt_nopun
```

```python
df['remove_Pun'] = df['Description'].apply(lambda x:remove_pun(x))
df['remove_Pun']
```

```python
#Tokenization

def tokenize(txt):
    tokens = re.split('\W+', txt)
    return tokens

df['clean_tok'] =  df['remove_Pun'].apply(lambda x: tokenize(x.lower()))

df['clean_tok']
```

```python
stopwords = nltk.corpus.stopwords.words('english') + nltk.corpus.stopwords.words('french')
print (stopwords)
```

```python
def remove_stopwords(txt_tokenized):
    txt_clean = [word for word in txt_tokenized if word not in stopwords]
    return txt_clean

df['no_stopW'] = df['clean_tok'].apply(lambda x: remove_stopwords(x))

df['no_stopW']
```

```python
#Detokenize 
text = df['no_stopW']

desc = []
for x in range(len(text)):
        s = TreebankWordDetokenizer().detokenize(text.iloc[x])
        desc.append(s)
desc
```

```python
#lemmatization
def lemmatization(desc, allowed_postags=["NOUN","ADJ","VERB","ADV"]):
 
    nlp = spacy.load("en_core_web_sm", disable=["parser","ner"])
    texts_out = []
    for desc in desc:
        doc = nlp(desc)
        new_text = []
        for token in doc:
            if token.pos_ in allowed_postags:
                new_text.append(token.lemma_)
        final = "".join(new_text)
        texts_out.append(new_text)
    return (new_text)


lemmatizaed_texts = lemmatization(desc)
lemmatizaed_texts
```

```python
def gen_words(desc):
    final=[]
    for desc in desc:
        new = gensim.utils.simple_preprocess(desc, deacc=True)
        final.append(new)
    return (final)

data_words = gen_words(desc)

data_words
```

```python
id2word = corpora.Dictionary(data_words)

corpus = []
for desc in data_words:
    new = id2word.doc2bow(desc)
    corpus.append(new)
    
print (corpus)
```

```python
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=3,
                                           random_state=100,
                                           update_every=1,
                                           chunksize=100,
                                           passes=10,
                                           alpha="auto")
```

```python
pyLDAvis.enable_notebook()
vis = gensimvis.prepare(lda_model, corpus, id2word)
vis
```
